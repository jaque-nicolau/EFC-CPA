##  Insertion Sort


```python
def insertionSort (arranjo):
  global checks
  for j in range (1,len(arranjo)):
    checks += 1
    chave = arranjo[j]
    i=j-1
    while i>=0 and arranjo[i] > chave:
      checks += 2
      arranjo[i+1] = arranjo[i]
      i=i-1
    arranjo[i+1] = chave
  return arranjo
```

##  Merge Sort


```python
def merge(A, aux, esquerda, meio, direita):
    global checks
    for k in range(esquerda, direita + 1):
        checks += 1
        aux[k] = A[k]
    i = esquerda
    j = meio + 1
    for k in range(esquerda, direita + 1):
        checks += 1
        if i > meio:
            checks += 1
            A[k] = aux[j]
            j += 1
        elif j > direita:
            checks += 1
            A[k] = aux[i]
            i += 1
        elif aux[j] < aux[i]:
            checks += 1
            A[k] = aux[j]
            j += 1
        else:
            A[k] = aux[i]
            i += 1
    


def mergeSort(A, esquerda=0, direita=None, aux = None):
    global checks
    if aux == None: aux = [0]*len(A)
    if direita == None: direita = len(A)-1
    if direita <= esquerda:
        checks += 1
        return
    meio = (esquerda + direita) // 2

    mergeSort(A, esquerda, meio, aux)
    mergeSort(A, meio + 1, direita, aux)
    merge(A, aux, esquerda, meio, direita)
```

##  Selection Sort


```python
def selectionSort(arranjo):
  global checks
  n = (len(arranjo)-1)
  for i in range(0,n):
    checks += 1
    valorMinimo = i
    for j in range (i+1, len(arranjo)):
      checks += 1
      if arranjo[j] < arranjo[valorMinimo]:
        checks += 1
        valorMinimo = j
    if valorMinimo != i:
      checks += 1
      temp = arranjo[i]
      arranjo[i] = arranjo[valorMinimo]
      arranjo[valorMinimo] = temp
  return arranjo
```

##  Bubble Sort


```python
def bubbleSort(arranjo):
  global checks
  n = len(arranjo)
  for i in range (n-1):
    checks += 1
    for j in range (0,n-i-1):
      checks += 1
      if arranjo[j] > arranjo [j+1]:
        checks += 1
        arranjo[j], arranjo[j+1] = arranjo[j+1], arranjo[j]
  return
```

## Quick Sort

> pivo = *A[A-comprimento]*

```python
def partition(values, left, right):
  global checks
  pivot = values[left]
  i = left
  j = left + 1
  while j <= right:
    checks += 1
    if values[j] <= pivot:
      checks += 1
      i+=1
      values[i], values[j] = values[j], values[i]
    j += 1
  values[i], values[left] = values[left], values[i]
  return i 
   

def quickSort(values, left=0, right=None):
  global checks
  if right == None: right = len(values)-1
  if left < right:
    checks += 1
    index_pivot = partition(values, left, right)
    quickSort(values, left, index_pivot - 1)
    quickSort(values, index_pivot + 1, right)	
  
```

#Funções auxiliares

```python
from collections import deque

def swap (A, i, j):
  temp = A[i]
  A[i] = A[j]
  A[j] = temp

def _partition(a, start, end):
  global checks
  pivot = a[end]
  pIndex = start
  for i in range(start, end):
    checks += 1
    if a[i] <= pivot:
      checks += 1
      swap(a, i, pIndex)
      pIndex = pIndex + 1
  swap(a, pIndex, end)
  return pIndex

def iterativeQuicksort(a):
  global checks
  stack = deque()
  start = 0
  end = len(a) - 1
  stack.append((start, end))
  while stack:
    checks += 1
    start, end = stack.pop()
    pivot = _partition(a, start, end)
    if pivot - 1 > start:
      checks += 1
      stack.append((start, pivot - 1))
    if pivot + 1 < end:
      checks += 1
      stack.append((pivot + 1, end))
```

```python
def execTime(func, arr):
  start = time.time_ns()
  func(arr)
  return (time.time_ns() - start) / 1000
```

```python
sizes = list(range(5, 1000, 5))
```

```python
def createArrays(func):
  arrays = []
  for size in sizes:
    a = []
    for i in range(size):
      a.append(func(i, size))
    arrays.append(a)
  return arrays
```

```python
def orderedArrays():
  return createArrays(lambda i, size: i)

def reversedArrays():
  return createArrays(lambda i, size: size - i)

def randomArrays():
  return createArrays(lambda i, size: random.randint(0, size))
```

```python
sortMethods = {
  'InsertionSort': {'func': insertionSort},
  'SelectionSort': {'func': selectionSort},
  'BubbleSort': {'func': bubbleSort},
  'HeapSort': {'func': heapSort},
  'MergeSort': {'func': mergeSort},
  'QuickSort': {'func': iterativeQuicksort},
}

newArrays = {'Ordered Array': orderedArrays, 'Reversed Array': reversedArrays, 'Random Array': randomArrays}

sortTypes = newArrays.keys()
```

```python
for sortMethod in sortMethods.values():
  for sortType in sortTypes:
    arrays = newArrays[sortType]()
    timeValues = []
    checkValues = []
    for array in arrays:
      checks = 0
      timeValues.append(execTime(sortMethod['func'], array))
      checkValues.append(checks)
    sortMethod[sortType] = timeValues
    sortMethod[sortType+'checks'] = checkValues
```

```python
def partitionRandom(nums, l, r):
    pivotIndex = random.randrange(l,r)
    nums[pivotIndex], nums[r] = nums[r], nums[pivotIndex]
    pivot, ptr = nums[r], l
    for i in range(l, r):
        if nums[i] <= pivot:
            nums[i], nums[ptr] = nums[ptr], nums[i]
            ptr += 1
    nums[ptr], nums[r] = nums[r], nums[ptr]
    return ptr
 

def quickSortRandom(nums, l=0, r=None):
    if r == None: r = len(nums)-1
    if len(nums) == 1:  
        return nums
    if l < r:
        pi = partitionRandom(nums, l, r)
        quickSortRandom(nums, l, pi-1)
        quickSortRandom(nums, pi+1, r)
    return nums

a = [9,2,4,8,5,2,8,5,0,9]
quickSortRandom(a)
a
 
```
